```julia
#Parameters
const g = 9.81
const N = 2 #number of links

#All links are identical
const m = 1.0 #link mass
const ℓ = 1.0 #link length
const J = 1.0 #link inertia
```

```julia
#Kinematics of an N-link planar pendulum
#Assume all joint angles are defined relative to the previous link

function kinematics(q)
    #Input: q is an N-dimensional vector of joint angles
    #Output: r is 3×N dimensional vector of global x,y,θ coordinates of each link
    
    r = zeros(eltype(q),3,N)
    
    #Assume first link is pinned to the origin
#     r[:,1] .= [0.5*ℓ*sin(q[1]); -0.5*ℓ*cos(q[1]); q[1]]
    r[:,1] .= [ℓ*sin(q[1]); -ℓ*cos(q[1]); q[1]]
    
    for k = 2:N
        θ = r[3,k-1] + q[k]
        x = r[1,k-1] + 0.5*ℓ*sin(r[3,k-1]) + 0.5*ℓ*sin(θ)
        y = r[2,k-1] - 0.5*ℓ*cos(r[3,k-1]) - 0.5*ℓ*cos(θ)
        
        r[:,k] .= [x;y;θ]
    end
    
    #output is r stacked as a single vector
    return r[:]
end
```

```julia
function potential(q)
    #Input: q is an N-dimensional vector of joint angles 
    #Output: U is potential energy
    
    r = reshape(kinematics(q),3,N)
    y = r[2,:]
    U = sum(m*g*y)
    
    return U
end
```

```julia
function kinetic(q,q̇)
    #Input: q is an N-dimensional vector of joint angles, q̇ is an N-dimensional vector of joint velocities
    #Output: T is kinetic energy
    
    K = ForwardDiff.jacobian(kinematics,q)
    ṙ = reshape(K*q̇,3,N)
    T = 0.5*m*(ṙ[1:2,1]'*ṙ[1:2,1]) + 0.5*J*ṙ[3,1]*ṙ[3,1]
    for k = 2:N
        T = T + 0.5*m*(ṙ[1:2,k]'*ṙ[1:2,k]) + 0.5*J*ṙ[3]*ṙ[3]
    end
    
    return T
end
```

```julia
function Lagrangian(q,q̇)
    return kinetic(q,q̇) - potential(q)
end
```

```julia
#These can be given
function D1L(q,q̇)
    return ForwardDiff.gradient(dq->Lagrangian(dq,q̇),q)
end

function D2L(q,q̇)
    return ForwardDiff.gradient(dq̇->Lagrangian(q,dq̇),q̇)
end

function D2D2L(q,q̇)
    return ForwardDiff.jacobian(dq̇->D2L(q,dq̇),q̇)
end

function D1D2L(q,q̇)
    return ForwardDiff.jacobian(dq->D2L(dq,q̇),q)
end
```

```julia
function M(q)
    #Input: q is an N-dimensional vector of joint angles
    #Output: M is an N×N positive-definite mass matrix
    
    return D2D2L(q,zeros(N))
end
```

```julia
function C(q,q̇)
    #Input: q is an N-dimensional vector of joint angles
    #Output: C is an N-dimensional coriolis force vector
    
    D1T = ForwardDiff.gradient(dq->kinetic(dq,q̇),q)
    return D1D2L(q,q̇)*q̇ - D1T
end
```

```julia
function G(q)
    #Input: q is an N-dimensional vector of joint angles
    #Output: G is an N-dimensional gravitational force vector
    
    return ForwardDiff.gradient(dq->potential(dq),q)
end
```

```julia
function manipulator_dynamics(x)
    #Input: x is an 2N-dimensional vector [q; q̇]
    #Output: ẋ is an 2N-dimensional vector [q̇; q̈]
    
    q = x[1:N]
    q̇ = x[N+1:end]
    
    q̈ = M(q)\(-C(q,q̇)-G(q))
    
    ẋ = [q̇; q̈]
end
```

```julia
# x0 = randn(2*N)
x0 = [deg2rad(85);0;0;0]
```

```julia
#Reference Solution

function f(x,p,t)
    manipulator_dynamics(x)
end

Tf = 30.0
tspan = (0.0,Tf)
prob = ODEProblem(f,x0,tspan)
sol = solve(prob,Tsit5(),abstol=1e-8,reltol=1e-8);
```

```julia
#Simulate with RK4

h = 0.2
T = 0:h:Tf
Nt = length(T)
x1 = zeros(2*N,Nt)

x1[:,1] .= x0

for k = 1:(Nt-1)
    ẋ1 = manipulator_dynamics(x1[:,k])
    ẋ2 = manipulator_dynamics(x1[:,k] + 0.5*h*ẋ1)
    ẋ3 = manipulator_dynamics(x1[:,k] + 0.5*h*ẋ2)
    ẋ4 = manipulator_dynamics(x1[:,k] + h*ẋ3)
    
    x1[:,k+1] .= x1[:,k] + (h/6.0)*(ẋ1 + 2*ẋ2 + 2*ẋ3 + ẋ4)
end
```

```julia
#Midpoint Discrete Lagrangian
function Ld(qk,qn)
    #Input: qk is an N-dimensional vector of joint angles at time k, qn is an N-dimensional vector of joint angles at time k+1
    #Output: Ld is a scalar discrete Lagrangian
    
    qm = 0.5*(qk+qn)
    vm = (qn-qk)/h
    return h*Lagrangian(qm,vm)
end
```

```julia
#We can provide these

function D1Ld(qk,qn)
    return ForwardDiff.gradient(dq->Ld(dq,qn),qk)
end

function D2Ld(qk,qn)
    return ForwardDiff.gradient(dq->Ld(qk,dq),qn)
end

function D2D1Ld(qk,qn)
    return ForwardDiff.jacobian(dq->D1Ld(qk,dq),qn)
end
```

```julia
function variational_midpoint_step(x)
    #Input: x is an 2N-dimensional vector [qk; q̇k]
    #Output: xn is an 2N-dimensional vector [qn; q̇n]
    
    q = x[1:N]
    q̇ = x[N+1:end]
    
    #Momentum
    p = M(q)*q̇
    
    qn = deepcopy(q)
    r = p + D1Ld(q,qn) #residual
    
    #Newton's method
    while maximum(abs.(r)) > 1e-12
        qn = qn - D2D1Ld(q,qn)\r
        r = p + D1Ld(q,qn)
    end
    
    #Calculate new momentum + velocity
    pn = D2Ld(q,qn)
    vn = M(qn)\pn
    
    xn = [qn; vn]
    
    return xn
end
```

```julia
#Simulate with Variational Midpoint
x2 = zeros(2*N,Nt)
x2[:,1] = x0

for k = 1:(Nt-1)
    x2[:,k+1] .= variational_midpoint_step(x2[:,k])
end
```
